```python
with open("dfiles.txt", "w", encoding="utf-8") as file:
# we create and open the file with "w"
    file.write('This is the first line of my text file. ')
    # writes str data into file
with open("dfiles.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the content
```

```python
with open("dfiles.txt", "a", encoding="utf-8") as file:
# we will add something to txt file with "a"
    file.write('This is the second line of my text file. ')
    # writes str data into file
with open("dfiles.txt", 'r', encoding="utf-8") as file:
    print(file.read())  # reads the content
```

```python
with open("d2files.txt", "w", encoding="utf-8") as file:
    file.write('My first sentence')
    file.write('My second sentence')
    file.write('My third sentence\n')
    file.write('My fourth sentence ')
    file.write('My last sentence')
with open("d2files.txt", 'r', encoding="utf-8") as file:
    print(file.read())
```

```python
fruits = ['Banana', 'Orange', 'Apple', 'Strwberry', 'Cherry']

with open("fruits.txt", "w", encoding='utf-8') as my_file :
    for basket in fruits :
        my_file.write(basket + "\n")
        
with open("fruits.txt", "r") as my_file :
    print(my_file.read())
```

```python
with open("istiklal.txt", "r", encoding="utf-8") as file:
   lines = file.readlines()
lines
```

```python
counter = 0
with open("istiklal.txt", "w", encoding="utf-8") as file:
    for i in lines :
        counter += 1
        if counter % 4 == 0:
            file.write(i + " \n")
        else :
            file.write(i)
            
with open("istiklal.txt", "r", encoding="utf-8") as file:
   print(file.read())
```
